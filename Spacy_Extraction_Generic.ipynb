```python
import spacy
import re
from spacy import displacy
```

```python
nlp1 = spacy.load('Three_entity_trained_model')
nlp = spacy.load('en_core_web_lg')
```

```python
def find_entities(contract_text):

    # applying pre_trained model

    result_pretrained = []
    sentences = contract_text.split('. ')
    for sent in sentences:
        processed_sent = nlp1(sent)
        
        for ent in processed_sent.ents:
            subject = sent.split(ent.text, 1)[0] + ' ' + ent.text + ' ' + ' '.join(sent.split(ent.text, 1)[1].split()[:6])
            
            if ent.text != 'LIBOR01':
                if ent.label_ == 'rate convention':
                    result_pretrained.append(('RATE CONVENTION', ent.text, sent))
                else:
                    pass
            else:
                pass

    # applying default model

    processed_text = nlp(contract_text)
    
    

    result = []
    for sent in processed_text.sents:
        processed_sent = nlp(sent.string)
        
        #displacy.render(processed_sent, style="ent", jupyter= True)
        #displacy.render(processed_sent, style="dep")

        spans = list(processed_sent.ents) + list(processed_sent.noun_chunks)
        for span in spans:
            span.merge()

        for ent in filter(lambda w: w.ent_type_ in ['MONEY', 'PERCENT', 'DATE', 'ORG'], processed_sent):

            n = list(re.finditer(ent.text, sent.string))[-1].start()
            subject = ' '.join(sent.string[:n].split()[-12:]) + ' ' + ent.text


            if ent.ent_type_ == 'MONEY':
                sign_container = [child for child in ent.children if child.dep_ == 'nmod']
                
                if len(sign_container) > 0:
                    sign = [child for child in ent.children if child.dep_ == 'nmod'][0]

                    result.append((ent.ent_type_, sign.string + ent.text, sent.string))

            elif ent.ent_type_ == 'DATE':

                date_children = list(filter(lambda x: x.ent_type_ == 'DATE', ent.children))
                if len(date_children) > 0:
                    children_str = ''
                    for child in date_children:
                        children_str += ', ' + child.string.strip()
                    result.append((ent.ent_type_, ent.text + children_str, subject))

                else:
                    if ent.dep_ == 'pobj' and ent.head.dep_ == 'prep' and 'the' not in ent.text.lower():
                        
                        if ent.head.head.dep_ == 'conj':
                            result.append((ent.ent_type_, ent.head.head.string + ent.head.string + ent.text, subject))
                        else:
                            result.append((ent.ent_type_, ent.text, subject))

                    elif ent.dep_ == 'advmod' and ent.head.dep_ != 'acomp':
                        result.append((ent.ent_type_, ent.text, subject))
                    
                    elif ent.dep_ == 'npadvmod':
                        result.append(('DURATION', ent.text, subject))

                    else:
                        pass

            elif ent.ent_type_ == 'ORG':

                exceptions = ['reuters', 'bloomberg', 'floating rate notes']
                if not any(exception in ent.text.lower() for exception in exceptions):
                    result.append((ent.ent_type_, ent.text, sent.string))
                else:
                    pass

            else:
                result.append((ent.ent_type_, ent.text, sent.string))

    unique_result = list(set(result))
    return unique_result + result_pretrained
```

```python
results = find_entities(master_text)
```
